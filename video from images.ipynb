# Create video from numpy array and images


# Video from numpy array

```python
import numpy as np
import cv2
import os
```

```python
height = 1280
width = 720
channel = 3
frame_per_sec = 30
sec = 5    # 20 sec
print("Video height-> ",height)
print("Video width-> ",width)
print("Video channel(RGB color)-> ",channel)
print("frame_per_sec-> ",frame_per_sec)
print("Video second-> ",sec)
```

```python
fourcc = cv2.VideoWriter_fourcc(*'MP42')
videos = cv2.VideoWriter('videos/demo.mkv', fourcc, float(frame_per_sec), (width,height))
print("Object->",videos)
```

```python
for frame_count in range(frame_per_sec*sec):
    img = np.random.randint(0,255, (height, width, channel), dtype = np.uint8)
    print("Image Matrix -> ",img)
    videos.write(img)
 
videos.release()
```

# Video from images


```python
height = 1280
width = 720
channel = 3
frame_per_sec = 10
sec = 20    # 5 sec

print("Video height-> ",height)
print("Video width-> ",width)
print("Video channel(RGB color)-> ",channel)
print("frame_per_sec-> ",frame_per_sec)
print("Video second-> ",sec)
```

```python
four_cc = cv2.VideoWriter_fourcc(*'MP42')
video_img = cv2.VideoWriter('videos/demo_images.mkv', four_cc, float(frame_per_sec), (width,height))
print("Object->",video_img)
```

```python
path = 'new_created_directory'
image_list = os.listdir(path)
print(image_list)
```

```python
for frame in range(frame_per_sec*sec):
    # select images randomly
    image_name = np.random.choice(image_list)
    print("Ranodmly Select image name -> ",image_name)
    # here join path + image
    img_path = os.path.join(path + '/' +image_name)
    print("Image full path -> ",img_path)
    # read image
    images = cv2.imread(img_path)
    print(images)
    image_resize = cv2.resize(images, (width,height))
    
    video_img.write(image_resize)
    
video_img.release()
```
