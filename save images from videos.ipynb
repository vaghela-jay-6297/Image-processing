# How to save images or numpy array as image

```python
# random numbers generate b/w 0 to 255
# width=300,height=300,channel=3(RGB color)
random_array = np.random.randint(255,size=(300,300,3)) 
response = cv2.imwrite('images/Random_image.png',random_array)
if response:
    print('Image successfully saved !')
else:
    print('Not found !')
        
```

```python
fetch_img = cv2.imread('images/Random_image.png')
cv2.imshow('Random images', fetch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Save images from videos


```python
# make new directory
os.mkdir('new_created_directory')
    
cap_video = cv2.VideoCapture('videos/short story.mp4')

count_img = 1

while cap_video.isOpened():
    res, frame = cap_video.read()
    
    if not res:
        print('Unable to read frame !!! ')
        break
        
    image_path_name = f"new_created_directory/created_image{ count_img }.png"
    
    make_image = cv2.imwrite(image_path_name, frame)
    if make_image:
        print(f"Image save at new_created_directory/created_image{ count_img }.png")
        count_img = count_img + 1 
    
    cv2.imshow('video',frame)
    
    if cv2.waitKey(25) & 0xff == ord('q'):
        break
              
cap_video.release()
cv2.destroyAllWindows()              
        
```
